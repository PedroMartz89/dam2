### Estad铆sticas b谩sicas



\\[
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
\quad \quad \quad \quad \quad
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})^2
\\]


```python
data = np.array([1, 2, 3, 4, 5])
print("Media:", np.mean(data))
print("Desviaci贸n est谩ndar:", np.std(data))
```

C谩lculo de la covarianza (por defecto usa estimador sesgado, se puede modificar usando el parametro ddof)
\\[
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
\\]

```python
# Datos de ejemplo (4 muestras, 3 caracter铆sticas)
X = np.array([[2.5, 2.4],
              [0.5, 0.7],
              [2.2, 2.9],
              [1.9, 2.2],
              [3.1, 3.0],
              [2.3, 2.7],
              [2, 1.6],
              [1, 1.1],
              [1.5, 1.6],
              [1.1, 0.9]])


# 2. Calcular la matriz de covarianza
cov_matrix = np.cov(X, rowvar=False)
cov_matrix
```

Cuando rowvar=False, significa que cada columna representa una variable y cada fila representa una observaci贸n (muestra o registro).

#### Operaciones matriciales

\\[
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
\times
\begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix}
=
\begin{pmatrix}
(1 \cdot 5 + 2 \cdot 7) & (1 \cdot 6 + 2 \cdot 8) \\
(3 \cdot 5 + 4 \cdot 7) & (3 \cdot 6 + 4 \cdot 8)
\end{pmatrix}
=
\begin{pmatrix}
19 & 22 \\
43 & 50
\end{pmatrix}
\\]

```python
# Crear dos matrices
matriz_a = np.array([[1, 2], [3, 4]])
matriz_b = np.array([[5, 6], [7, 8]])

# Multiplicaci贸n de matrices (producto punto)
multiplicacion_matrices = np.dot(matriz_a, matriz_b)
print(multiplicacion_matrices)
```
