EJERCICIO 2 - CALCULA LA MEDIA DE CADA ALUMNO Y DE LA CLASE.

```python

import pandas as pd
import numpy as np
alumnos = pd.Series(['alu1','alu2','alu3','alu4','alu5','alu6','alu7','alu8'])

ra1 = pd.Series([5, 3, 7, 6, 7, 2, 3, 5])
ra2 = pd.Series([6, 2, 8, 4, 8, 6, 7, 6])
ra3 = pd.Series([5, 5, 6, 5, 9, 9, 8, 7])
ra4 = pd.Series([7, 3, 8, 5, 4, 4, 9, 6])
ra5 = pd.Series([4, 1, 9, 6, 10, 7, 6, 5])

df = pd.DataFrame({'Alumno':alumnos, 'RA1':ra1, 'RA2':ra2, 'RA3': ra3, 'RA4': ra4, 'RA5':ra5})

df['MEDIA'] = (df['RA1'] + df['RA2'] + df['RA3'] + df['RA4'] + df['RA5']) / 5

print(df)

print("Media de clase" , df['MEDIA'].mean())
```

EJERCICIO 7: Añade 2 alumnos a la estructura de datos generando sus calificaicones de forma aleatoria con valores entre 1 y 9



```python
import numpy as np
import pandas as pd

alumnos = pd.Series(['alu1','alu2','alu3','alu4','alu5','alu6','alu7','alu8'])
ra1 = pd.Series([5, 3, 7, 6, 7, 2, 3, 5])
ra2 = pd.Series([6, 2, 8, 4, 8, 6, 7, 6])
ra3 = pd.Series([5, 5, 6, 5, 9, 9, 8, 7])
ra4 = pd.Series([7, 3, 8, 5, 4, 4, 9, 6])
ra5 = pd.Series([4, 1, 9, 6, 10, 7, 6, 5])

df = pd.DataFrame({
    'Alumno': alumnos,
    'RA1': ra1,
    'RA2': ra2,
    'RA3': ra3,
    'RA4': ra4,
    'RA5': ra5
})

nuevo1 = np.random.randint(1, 10, size=5)
nuevo2 = np.random.randint(1, 10, size=5)

df_nuevos = pd.DataFrame({
    'Alumno': ['alu9', 'alu10'],
    'RA1': [nuevo1[0], nuevo2[0]],
    'RA2': [nuevo1[1], nuevo2[1]],
    'RA3': [nuevo1[2], nuevo2[2]],
    'RA4': [nuevo1[3], nuevo2[3]],
    'RA5': [nuevo1[4], nuevo2[4]]
})


df = pd.concat([df, df_nuevos], ignore_index=True)

print(df)
```

EJERCICIO 8 (2 puntos): Para superar el módulo es necesario que los alumos tengan superados todos los RA. Obten un listado de aquellos alumnos que no han superado el módulo junto con su nota media, para ello debes añadir una columna nueva que con los textos "Superado" "NoSuperdo" indique si el alumno ha superado la materia.

```python
df['MEDIA'] = df[['RA1','RA2','RA3','RA4','RA5']].mean(axis=1)
df['SUPERADO'] = df[['RA1','RA2','RA3','RA4','RA5']].min(axis=1) >= 5
df['SUPERADO'] = df['SUPERADO'].map({True:'Superado', False:'NoSuperado'})

print(df[df['SUPERADO'] == 'NoSuperado'][['Alumno', 'MEDIA', 'SUPERADO']])
```
