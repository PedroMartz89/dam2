2. Operadores Aritm칠ticos
Adem치s de los b치sicos, Python incluye operadores 칰tiles para algoritmos.

```python
a, b = 13, 4

print("Suma:", a + b)
print("Divisi칩n Real:", a / b)   # Devuelve decimales (3.25)
print("Divisi칩n Entera:", a // b) # Descarta decimales (3)
print("M칩dulo (Resto):", a % b)   # 칔til para saber si un n칰mero es par (a % 2 == 0)
print("Potencia:", a ** 2)        # 13 al cuadrado
```

3. Condicionales (Toma de decisiones)

La clave en Python es la indentaci칩n (el sangrado). Todo lo que est칠 "dentro" del if debe llevar 4 espacios.

```python
nota = 8.5

if nota >= 9:
    print("Excelente")
elif nota >= 5:
    print("Aprobado")
else:
    print("Suspenso")
```

4. Estructuras de Datos: Listas, Tuplas y Diccionarios

```python
# --- LISTA: Ordenada y se puede modificar (MUTABLE) ---
alumnos = ["Ana", "Luis"]
alumnos.append("Marta") # A침adir al final
alumnos[0] = "Adriana"  # Modificar el primer elemento

# --- TUPLA: Ordenada pero NO se puede modificar (INMUTABLE) ---
# Se usa para datos que deben ser constantes o protegidos.
coordenadas = (40.41, -3.70)

# --- DICCIONARIO: Colecci칩n de par Clave: Valor ---
usuario = {"id": 1, "username": "python_user"}

print(f"Lista: {alumnos}")
print(f"Tupla: {coordenadas}")
print(f"Nombre de usuario: {usuario['username']}")
```

5. Bucles (Iteraciones)

El bucle for en Python es muy potente para recorrer colecciones directamente.

```python
# Recorrer una lista
print("Lista de alumnos:")
for alumno in alumnos:
    print(f"- {alumno}")

# Bucle con rango (range)
print("\nContando del 0 al 2:")
for i in range(3):
    print(i)
```

6. Cadenas de Texto (Strings) 游댟

Las cadenas son secuencias de caracteres inmutables. Esto significa que no puedes cambiar una letra directamente (ej: nombre[0] = 'A' dar치 error), debes crear una nueva cadena.

```python
# M칠todos de las cadenas
mensaje = "  aprender python es divertido  "

# Formatear el texto
print(f"Original: '{mensaje}'")
# Covertir en may칰sculas
print(f"May칰sculas: {mensaje.upper()}")
# Elimina los espacios en blanco
print(f"Sin espacios laterales: '{mensaje.strip()}'")
# Busca una subcadena (par치metro 1) y la remplaza (par치metro 2)
print(f"Reemplazar: {mensaje.replace('divertido', 'genial')}")

# Dividir la cadena en una lista
palabras = mensaje.split()
print(f"Lista de palabras: {palabras}")
```

**Iteraci칩n sobre Cadenas**

Podemos recorrer cada car치cter utilizando un bucle for.

```python
palabra = "Python"
print("Iterando sobre la palabra:")
for letra in palabra:
    print(f"Letra actual: {letra}")
```
